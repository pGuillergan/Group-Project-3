```python
%matplotlib inline
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
mergedf = pd.read_csv('metadata_otu_merged.csv')
mergedf.head(1)
```

```python
mergedf = mergedf.drop("Unnamed: 0", axis=1)
```

```python
print(mergedf.shape)
```

```python
# print(mergedf.iloc[:,[1085]])
```

```python
X = mergedf.iloc[:, 52:1085]
y = mergedf["sample_type"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
# y_scaler = StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# y_train_scaled = y_scaler.transform(y_train)
# y_test_scaled = y_scaler.transform(y_test)
```

```python
#decision tree non-scaled
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
#from non-scaled data
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train.ravel())
rf.score(X_test, y_test) #geni impurity coefficient
```

```python
# from scaled data
rf_scaled = RandomForestClassifier(n_estimators=200)
rf_scaled = rf_scaled.fit(X_train_scaled, y_train.ravel())
rf_scaled.score(X_test_scaled, y_test) #geni impurity coefficient
```
