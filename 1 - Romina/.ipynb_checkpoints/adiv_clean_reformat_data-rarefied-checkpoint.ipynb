```python
import pandas as pd
import numpy as np
import os
```

```python
adiv = pd.read_csv("adiv_chao1_even5000.csv")

adiv.head()
```

```python
metadata = pd.read_csv("../data/metadata_CSV.csv")
metadata.head(1)
```

```python
dog_fam = pd.read_csv("dogFamily_toMerge.csv")
```

```python
print(adiv.shape)
print(metadata.shape)
adiv = metadata.merge(adiv, left_on="sample_name", right_on="sample_name", suffixes=(False,False))
# Traceback (most recent call last):
# ...
# ValueError: columns overlap but no suffix specified:
#     Index(['value'], dtype='object')
print(adiv.shape)
adiv
```

```python
#MERGE with DOG_FAM info

adiv = adiv.merge(dog_fam, left_on="family", right_on="family", suffixes=(False,False), how="outer")
adiv["have_dog"].fillna("no",inplace = True)
print(adiv.shape)
adiv.head()
```

```python
adiv.to_csv('adiv_5000_merged.csv')
```
